```python
updated_log_df = log_df.copy()
updated_log_df['User_Agent'] = updated_log_df['User_Agent'].replace('-', np.nan)
updated_log_df['Referrer'] = updated_log_df['Referrer'].replace('-', np.nan)
```

```python
import re

updated_log_df['dotfile_access'] = updated_log_df['Path'].apply(
    lambda p: 1 if re.search(r'/\.[^/]+', p) else 0
)
```

```python
updated_log_df
```

```python
updated_log_df.drop_duplicates(subset = ['IP','Request_method','Path','Protocol','Response_code','Referrer','User_Agent','dotfile_access'],inplace=True)
```

```python
updated_log_df
```

```python
updated_log_df['bad_user_agent'] = updated_log_df['User_Agent'].str.lower().str.contains(
    r"(?:bot|curl|scraper|wget|httpclient|requests|spider|crawler|expanse|censys|modat)",
    na=True
).astype(int)
```

```python
updated_log_df
```

```python
updated_log_df['no_referrer'] = updated_log_df['Referrer'].apply(lambda r: 1 if r is np.nan else 0)
```

```python
updated_log_df['no_referrer'].value_counts()
```

```python
updated_log_df
```

```python
suspicious_keywords = ['wp-admin', 'phpmyadmin', 'config', 'admin', 'setup', 'shell', 'login', 'cmd', 'api', 'backup']
updated_log_df['suspicious_path'] = updated_log_df['Path'].str.lower().apply(
    lambda x: 1 if any(keyword in x for keyword in suspicious_keywords) is True else 0
)
```

```python
updated_log_df
```

```python
updated_log_df['is_bot'] = updated_log_df['dotfile_access'] | updated_log_df['bad_user_agent'] | updated_log_df['suspicious_path']
```

```python
updated_log_df['is_bot'].value_counts()
```

```python
from user_agents import parse
import pandas as pd
import numpy as np

# Function to extract parts from a user agent string, safely
def extract_ua_parts(ua_string):
    if pd.isna(ua_string):  # Check for NaN or missing values
        return pd.Series({
            'ua_browser': np.nan,
            'ua_os': np.nan,
            'ua_is_mobile': np.nan,
            'ua_is_pc': np.nan,
        })
    ua = parse(ua_string)
    return pd.Series({
        'ua_browser': ua.browser.family,
        'ua_os': ua.os.family,
        'ua_is_mobile': 1 if ua.is_mobile else 0,
        'ua_is_pc': 1 if ua.is_pc else 0,
    })

# Apply function
ua_features = updated_log_df['User_Agent'].apply(extract_ua_parts)

# Combine with the original DataFrame
updated_log_df= pd.concat([updated_log_df, ua_features], axis=1)


''' 
from user_agents import parse

ua = parse("Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36 Edg/90.0.818.46")
print(ua.os.family)       # e.g., 'Windows'
print(ua.browser.family)  # e.g., 'Chrome'
print(ua.device.family)   # e.g., 'Other' or 'iPhone'
'''
```

```python
top_5_browsers = ['Chrome','Firefox','Edge','Chrome Mobile','Opera']
top_operating_systems = ['Windows','iOS','Ubuntu','Linux','Mac OS X','Android']
updated_log_df['ua_browser'] = updated_log_df['ua_browser'].apply(lambda x : x if x in top_5_browsers else 'Other')
updated_log_df['ua_os'] = updated_log_df['ua_os'].apply(lambda x : x if x in top_operating_systems else 'Other')
```
